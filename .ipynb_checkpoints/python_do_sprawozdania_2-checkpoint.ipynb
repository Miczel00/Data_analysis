```python
dane_auto = pd.read_excel(r'C:\Users\Maja\Desktop\pliki do sprawozdania nr2\autoskalowane_catvsdog.xlsx')
dane_auto
```

```python
dane_autoskalowane = pd.DataFrame(dane_auto)
```

```python
autokskal = dane_autoskalowane.drop(['Lokalizacja'], axis = 1)
autokskal
```

```python
macierz_odleglosci_euklides = np.zeros((49,49))
```

```python
#2. Obliczanie odległości między obiektami: ODLEGŁOŚĆ EUKLIDESOWA
sum = 0
for elementA in range(len(dane_autoskalowane)):
    for elementB in range(len(dane_autoskalowane)):
        for attribute in range(1,9): #
            sum += (dane_autoskalowane.loc[elementA][attribute] - dane_autoskalowane.loc[elementB][attribute]) ** 2
        macierz_odleglosci_euklides[elementA][elementB] = sqrt(sum)
        sum = 0
```

```python
macierz_odl = pd.DataFrame(macierz_odleglosci_euklides)
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
#Single linkage liczone ręcznie
```

```python
lista_liczb = []
for i in range(len(macierz_odl), 200):
    lista_liczb.append(i)

iterator = iter(lista_liczb)

min_val_list = []
```

```python
lista_k = []
lista_wsp = []
min_val_list = []


while macierz_odl.size >= 4:
    
    macierz_odl_min = macierz_odl.replace(0, np.inf)
    min_val = macierz_odl_min.values.min()
    xy = np.where(macierz_odl_min == min_val)
    wspolrzedne = xy[0][0], xy[1][0]

    min_wsp = min(wspolrzedne)
    max_wsp = max(wspolrzedne)
    min_val_list.append(min_val)
    new_name = next(iterator)
    
    lista_k = macierz_odl.iloc[min_wsp].name, macierz_odl.iloc[max_wsp].name
    lista_wsp.append(lista_k)
    
    wektor = []
    for i in range(len(macierz_odl)):
        if macierz_odl.iloc[wspolrzedne[0], i] < macierz_odl.iloc[wspolrzedne[1], i]:
            wektor.append(macierz_odl.iloc[wspolrzedne[0], i])
        else:
            wektor.append(macierz_odl.iloc[wspolrzedne[1], i])

    macierz_odl.iloc[min_wsp] = wektor
    macierz_odl.iloc[:, min_wsp] = wektor

    macierz_odl.drop(macierz_odl.index[max_wsp], axis = 0, inplace = True)
    macierz_odl.drop(macierz_odl.columns[max_wsp], axis = 1, inplace = True)

    macierz_odl.rename(columns = {macierz_odl.columns[min_wsp] : new_name}, inplace = True)
    macierz_odl.rename(index = {macierz_odl.index[min_wsp] : new_name}, inplace = True)
```

```python
print(lista_wsp, min_val_list)
```

```python
#Ilość obiektów w klastrze
macierz = [1 for i in range(len(autokskal)*2)]
for k in range(len(autokskal) - 1):
    skład1 = lista_wsp[k][0]
    skład2 = lista_wsp[k][1]
    macierz[k + len(autokskal)] = macierz[skład1] + macierz[skład2]
    print(lista_wsp[k][0], lista_wsp[k][1], macierz[k + len(autokskal)])
```

```python
tabela_den = []
for i in range(len(autokskal) - 1):
    lista = []
    lista.append(lista_wsp[i][0])
    lista.append(lista_wsp[i][1])
    lista.append(min_val_list[i])
    lista.append(macierz[i + len(autokskal)])
    tabela_den.append(lista)
        
print(tabela_den)
```

```python
fig = plt.figure(figsize=(15, 10)) #figsize=(15, 100)
dn = hierarchy.dendrogram(tabela_den)
plt.xlabel("Objects")
plt.ylabel("Distances")
plt.title("Dendrogram \n(Euclidean distance, single-linkage method)\nManual counting")
plt.show()
```

```python
##3, 4. Analiza HCA, dendrogram: sprawdzenie wbudowaną funkcją
dendrogram_euklides_single = hierarchy.linkage(autokskal, method='single')
  
plt.figure(figsize=(15,10))
plt.xlabel("Objects")
plt.ylabel("Distances")
plt.title("Dendrogram \n(Euclidean distance, single-linkage method)")

dendrogram = hierarchy.dendrogram(dendrogram_euklides_single)
```

```python

dendrogram_euklides_ward = hierarchy.linkage(macierz_odleglosci_euklides, method='ward')
  
plt.figure(figsize=(15,7))
plt.xlabel("Objects")
plt.ylabel("Distances")
plt.title("Dendrogram \n(Euclidean distance, Ward's method)")

dendrogram = hierarchy.dendrogram(dendrogram_euklides_ward)
```

```python
dendrogram_euklides_single = hierarchy.linkage(macierz_odleglosci_euklides, method='complete')
  
plt.figure(figsize=(15,10))
plt.xlabel("Objects")
plt.ylabel("Distances")
plt.title("Dendrogram \n(Euclidean distance, complete-linkage method)")

dendrogram = hierarchy.dendrogram(dendrogram_euklides_single)
```

```python
macierz_odleglosci_manhattan = np.zeros((49,49))
```

```python
#Obliczanie odległości między obiektami: ODLEGŁOŚĆ MANHATTAN
s = 0
for elementA in range(len(dane_autoskalowane)):
    for elementB in range(len(dane_autoskalowane)):
        for attribute in range(1,9):
            a = (dane_autoskalowane.iloc[elementA,attribute] - dane_autoskalowane.iloc[elementB,attribute])
            s += abs(a)
        macierz_odleglosci_manhattan[elementA][elementB] = s
        s = 0
```

```python
manhattan = pd.DataFrame(macierz_odleglosci_manhattan)
manhattan
```

```python
manhattan.to_excel("macierz_odl_manhattan.xlsx")
```

```python
dendrogram_manhattan_ward = hierarchy.linkage(macierz_odleglosci_manhattan, method='ward')
  
plt.figure(figsize=(15,7))
plt.xlabel("Objects")
plt.ylabel("Distances")
plt.title("Dendrogram \n(Manhattan distance, Ward's method)")

dendrogram = hierarchy.dendrogram(dendrogram_manhattan_ward)
```

```python
dendrogram_manhattan_single = hierarchy.linkage(macierz_odleglosci_manhattan, method='single')
  
plt.figure(figsize=(15,9))
plt.xlabel("Objects")
plt.ylabel("Distances")
plt.title("Dendrogram \n(Manhattan distance, single-linkage method)")
  
dendrogram = hierarchy.dendrogram(dendrogram_manhattan_single)
```

```python
dendrogram_manhattan_complete = hierarchy.linkage(macierz_odleglosci_manhattan, method='complete')
  
plt.figure(figsize=(15,6))
plt.xlabel("Objects")
plt.ylabel("Distances")
plt.title("Dendrogram \n(Manhattan distance, complete-linkage method)")
  
dendrogram = hierarchy.dendrogram(dendrogram_manhattan_complete)
```

```python
#Obliczanie odległości między obiektami: ODLEGŁOŚĆ CZEBYSZEWA
```

```python
macierz_odleglosci_czebyszew = np.zeros((49,49))
```

```python
s = []
for elementA in range(len(dane_autoskalowane)):
    for elementB in range(len(dane_autoskalowane)):
        for attribute in range(1,9):
            a = abs(dane_autoskalowane.iloc[elementA,attribute] - dane_autoskalowane.iloc[elementB,attribute])
            s.append(a)
        macierz_odleglosci_czebyszew[elementA][elementB] = max(s)
        s = []
```

```python
czebyszew = pd.DataFrame(macierz_odleglosci_czebyszew)
czebyszew
```

```python
czebyszew.to_excel("macierz_odl_czebyszew.xlsx")
```

```python
dendrogram_czebyszew_ward = hierarchy.linkage(macierz_odleglosci_czebyszew, method='ward')
  
plt.figure(figsize=(15,7))
plt.xlabel("Objects")
plt.ylabel("Distances")
plt.title("Dendrogram \n(Chebyshev distance, Ward's method)")
  
dendrogram = hierarchy.dendrogram(dendrogram_czebyszew_ward)
```

```python
dendrogram_czebyszew_single = hierarchy.linkage(macierz_odleglosci_czebyszew, method='single')
  
plt.figure(figsize=(15,6))
plt.xlabel("Objects")
plt.ylabel("Distances")
plt.title("Dendrogram \n(Chebyshev distance, single-linkage method)")

dendrogram = hierarchy.dendrogram(dendrogram_czebyszew_single)
```
